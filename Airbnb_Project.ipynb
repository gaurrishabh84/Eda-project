## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

### Chart - 1
**Top 10 Host With Most Listings**


```python
# Chart - 1 visualization code
top_host = airbnb.host_id.value_counts().head(10)
top_host_df = pd.DataFrame(top_host)
top_host_df.reset_index(inplace=True)
top_host_df.rename(columns={'index':'HOST_ID','host_id':'LISTING_COUNT'}, inplace=True)

sns.set(rc={'figure.figsize':(10,6)})
sns.set_style('white')
plot_host = sns.barplot(x='HOST_ID', y='LISTING_COUNT', data=top_host_df, palette='crest')
plot_host.set_title('Host With Most Listings In NYC')
plot_host.set_xlabel('Host Id',)
plot_host.set_ylabel('Count of Listings')
plot_host.tick_params(axis='x', rotation=60)
```

##### 1. Why did you pick the specific chart?

Because Bar Plot is supposed to show the magnitude of each data point and the proportions between all the data points.

##### 2. What is/are the insight(s) found from the chart?

These are the top 10 Host Id which have the most number of listings in the dataset.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes,this insights helps in customer experiencing for their bookings and also create a large business impact by showing these listings to their customers as more listings is directly proportional to availability.

### Chart - 8 - Pair Plot

```python
# Pair Plot visualization code
#grabbing top 10 neighbourhoods for sub-dataframe
sub_7 = airbnb.loc[airbnb['neighbourhood'].isin(['Williamsburg','Bedford-Stuyvesant','Harlem','Bushwick',
                 'Upper West Side','Hell\'s Kitchen','East Village','Upper East Side','Crown Heights','Midtown'])]

#using catplot to represent multiple interesting attributes together and a count
viz_3=sns.catplot(x='neighbourhood', hue='neighbourhood_group', col='room_type', data=sub_7, kind='count')
viz_3.tick_params(axis='x', rotation=60)
plt.show()
```

##### 1. Why did you pick the specific chart?

With the use of one of many visual representations, this function gives users access to a number of axes-level functions that illustrate the connection between numerical data and one or more category variables.

##### 2. What is/are the insight(s) found from the chart?

Manhattan and Brooklyn are one of the most traveled destinations, therefore would have the most listing availability. We can also observe that Bedford-Stuyvesant and Williamsburg are the most popular for Manhattan borough, and Harlem for Brooklyn.
